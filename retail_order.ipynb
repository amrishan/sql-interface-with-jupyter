```python
%%sql
with cte1 AS 
(
  select customer_id,
         sum(quantity * unit_price) AS spend,
         count(distinct invoice_number) AS freq
    from invoices
group by customer_id
 ),

cte2 AS
( 
  select customer_id, 
         spend,
         freq,
         dense_rank() over(order by spend desc) as rank
  from cte1
),

cte3 AS
( 
  select customer_id, 
         spend,
         freq,
         DENSE_RANK() OVER(order by freq desc) as rank
  from cte1
),

cte4 AS
( 
  select customer_id, 
         spend,
         freq,
         DENSE_RANK() OVER(order by spend desc, freq desc) as rank
  from cte1
)

select * from cte2 limit 3;
```

```python
%%sql

with cte1 as
(select
   customer_id,
   invoice_date
 from invoices
 group by customer_id, invoice_date
),
cte2 as(
  select 
    customer_id,
    invoice_date,
    lead(invoice_date, 1) over(partition by customer_id order by invoice_date) as next_inv_date
  from cte1
  ),
cte3 as(
  select 
    customer_id,
    avg(next_inv_date - invoice_date) as days_bw_purch
  from cte2
  where next_inv_date is not null
  group by customer_id)
  
select cast(avg(days_bw_purch) as float)  from cte3;
```

```python
%%sql
with cte1 as
(
  select 
    invoice_date,
    SUM(quantity * unit_price) AS spend
  from invoices
  group by invoice_date
),

cte2 as 
(
  select
      p.promotion_type,
      sum(cte1.spend)
  from cte1
  inner join promotions as p
  on p.promotion_date = cte1.invoice_date
  group by p.promotion_type
)

select * from cte2 order by total_spend desc;
```

```python
%%sql
with cte1 as
(
  select *
  from invoices as i
  inner join promotions as p
  on i.invoice_date = p.promotion_date
  inner join customers as c
  on i.customer_id = c.customer_id
)

select concat(cast((select count(distinct invoice_number) 
from cte1 where cte1.email_subscriber = 'f') as float)/ cast((select count(distinct invoice_number) from cte1) as float) * 100, '%') as perc_sub;
```

```python
%%sql
with cte1 as 
(
  select 
    i.invoice_date,
    sum(quantity * unit_price) as sale
  from invoices as i
  left join promotions as p
  on i.invoice_date = p.promotion_date
  where p.promotion_date is null
  group by i.invoice_date
)
  
select * from cte1 order by sale desc;
```
